# Data Preprocessing Tools

## Importing the libraries

```python
import pandas as pd # for importing data from files
import matplotlib.pyplot as plt  # used for plotting chart and graphs
import numpy as np  # used for manuplating data
```

## Importing the dataset

```python
data_set = pd.read_csv("Data.csv")
# csv is represented as a matrix 
# first specify the rows then the columns 

X = data_set.iloc[:,:-1].values # [specify_row(all),specify_column(all except the last)]
Y = data_set.iloc[:,-1].values
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')

imputer.fit(X[:,1:3]) # only fix column 1 and 2 i.e [1:3]
X[:,1:3] = imputer.transform(X[:,1:3])
```

## Encoding categorical data

### Encoding the Independent Variable

```python
# Representing variables interms of  binary values // encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[0])],remainder="passthrough")
X = np.array(ct.fit_transform(X)) # fit_transform doesn't return numpy array
```

### Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = np.array(le.fit_transform(Y))
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=1) 
```
