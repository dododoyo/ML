# Polynomial Regression

## Importing the libraries

```python
import numpy as np 
from matplotlib import pyplot as plt
import pandas as pd 
```

## Importing the dataset

```python
data_set = pd.read_csv("./Position_Salaries.csv").values

X = data_set[:,1:-1]
y = data_set[:,-1]

```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X,y)
```

## Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures

# create a model with degree of 2 
polynomial_option = PolynomialFeatures(degree=4)
polynomial_x = polynomial_option.fit_transform(X)

'''
The fit_transform method is a combination of two steps: fitting and transforming. First, the method fits the transformer to the data X. This means it learns the necessary parameters from the data, such as the coefficients for polynomial terms in the case of polynomial regression. After fitting, it immediately transforms the data X based on the learned parameters, producing a new array polynomial_x that contains the transformed features.
'''

polynomial_model = LinearRegression()
polynomial_model.fit(polynomial_x,y)
```

## Visualising the Linear Regression results

```python
plt.scatter(X,y,color='red')
plt.plot(X,linear_model.predict(X))
plt.title('Salary Plot- Linear')
plt.xlabel("Postion")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Polynomial Regression results

```python
plt.scatter(X, y, color='red')
plt.plot(X, polynomial_model.predict(polynomial_x))

plt.title('Salary Plot- Polynomial')
plt.xlabel("Postion")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
plt.scatter(X, y, color='red')
plt.plot(X, polynomial_model.predict(polynomial_x))

plt.title('Salary Plot- Polynomial')
plt.xlabel("Postion")
plt.ylabel("Salary")
plt.show()
```

## Predicting a new result with Linear Regression

```python
# predict salary for 6.5
linear_model.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```python
# predict salary for 6.5
polynomial_model.predict(polynomial_option.fit_transform([[6.5]]))
```
