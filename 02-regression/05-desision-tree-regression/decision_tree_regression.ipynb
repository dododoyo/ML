# Decision Tree Regression

## Importing the libraries

```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
```

## Importing the dataset

```python
data_set = pd.read_csv("./Position_Salaries.csv").values

X = data_set[:,1:-1]
y = data_set[:,-1]
```

## Training the Decision Tree Regression model on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor  

# while DecisionTreeRegressor creates a continous
# numerical value the DecisionTreeClassifier 
# will create a distinct Category

dtr_model = DecisionTreeRegressor(random_state=0)
dtr_model.fit(X,y)
```

## Predicting a new result

```python
y_model = dtr_model.predict([[6.5]])
```

## Visualising the Decision Tree Regression results

```python
plt.scatter(X,y,color='red')

y_model = dtr_model.predict(X)

plt.plot(X,y_model)

plt.title("Salary Prediction (Decision Tree Regression)")
plt.xlabel("Position")
plt.ylabel("Salary")

plt.show()
```

## Visualising the Decision Tree Regression results (higher resolution)

```python
X_grid = np.arange(np.min(X), np.max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(X,y, color="red")

y_prediction = dtr_model.predict(X_grid)

plt.plot(X_grid, y_prediction)

plt.title("Salary Prediction (SVR)")
plt.xlabel("Position Level")
plt.ylabel("Salary")

plt.show()
```
