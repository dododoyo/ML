## Contour Detection

### Import Libraries 

```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 as cv 
```

# Import Image 

```python
contour_image = cv.imread("../data/Files/DATA/internal_external.png",0)
# plt.imshow(contour_image,cmap='gray')
```

```python
contourResult = cv.findContours(contour_image,cv.RETR_CCOMP,cv.CHAIN_APPROX_SIMPLE)
```

```python
contours,heirarcy = contourResult
```

```python
# create an empty image to display the detected contours
external_contours = np.zeros(contour_image.shape)
```

### Displaying each contour independently

- contours array holds all the contours detected from the image.

> using a for loop we iterate over all the contours and paint only the ones we need based on their heirarchy.

```python
for i in range(len(contours)): 
    # External Contour
    if heirarcy[0][i][3] == -1:
        # draw the contours[i] on the blank page if 
        # the heirarcy is -1 
        cv.drawContours(external_contours,contours,i,255,-1)
plt.imshow(external_contours, cmap='gray')
```

- create new blank page to create a new image

```python
# create an empty image to display the detected contours
internal_contours = np.zeros(contour_image.shape)
for i in range(len(contours)):
    # Internal Contour
    if heirarcy[0][i][3] != -1:
        # draw the contours[i] on the blank page if
        # the heirarcy is -1
        cv.drawContours(internal_contours, contours, i, 255, -1)
plt.imshow(internal_contours, cmap='gray')
```

- Find out how many distinct contours there are.

```python
distinct_contours = set()

for h in heirarcy[0]:
    distinct_contours.add(h[3])

print(len(distinct_contours))
print(distinct_contours)
```
